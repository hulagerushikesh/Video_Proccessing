In this part we are going to make a video from the directory :::::::

```python
import os
from moviepy.editor import *
from PIL import Image

p1 = os.path.join('data')
# getting the path for the  inputs
p2 = os.path.join(p1 , 'samples')
path = os.path.join(p2 , 'inputs')
out_path = os.path.join(p2 , 'outputs')

thumbnails_dir = os.path.join(path , 'thumbnails')
os.makedirs(thumbnails_dir , exist_ok=True)

thumbnails_per_frame_dir = os.path.join(path , 'thumbnails_per_frame')
os.makedirs(thumbnails_per_frame_dir , exist_ok=True)
```

```python
# Now just give the path for the output video :
out_video = os.path.join(out_path , 'my_video.mp4')
```

```python
# Create the image sequence :
this_dir = os.listdir(thumbnails_dir)
# Get the images from the gievn directory :
# filepath =  []
# for fname in this_dir :
#     if fname.endswith('jpg'):
#         path_dir = os.path.join(thumbnails_dir , fname)
#         filepath.append(path_dir)
        
        
# There is also an way to create the directory path using list comprehenssions :
filepath_1 = []
filepath_1 = [os.path.join(thumbnails_dir, fname) for fname in this_dir if fname.endswith('jpg')]
```

```python
# Crate an video :
clip = ImageSequenceClip(filepath_1 , fps = 20)
clip.write_videofile(out_video)
```

```python
# But there is a problem , and it is about the reordering of the images. As so far our aim is not complete yet 
# we want to make an appropriate video having correct sequence of the images 

# Lets just check how the sequence unordered (just by printing filepath_1 which holds the image sequence) :

print(filepath_1)
```

```python
# Ohhh its clearly seen that the unordered sequence : 
# 0 -> 1 -> 11 ->    and somewhere after 19 -> 2 
# So this is very inconvinient filepath has been created , we need to sort this image sequence and goinh to make a new video.
```

```python
# So here we going to use the hardest way to traverse through a image sequence using 'rock' :

# create the dict :  To make a linear format of the sequence -->  by making a key value pair :

dict_dir = {}
for root , dirs , files in os.walk(thumbnails_per_frame_dir):
    for fname in files:
        filepath_2 = os.path.join(root , fname)
        try:
            key = float(fname.replace('.jpg' , ''))
        except:
            key = None
        if(key != None) :
            dict_dir[key] = filepath_2
    
            
```

```python
# So now the images are in order , lets create a path for images and from that make a new  video :

new_paths = []
for k in sorted(dict_dir.keys()):
    filepath_2 = dict_dir[k]
    new_paths.append(filepath_2)
    
# Now just give the path for the output video (the new one):
out_video2 = os.path.join(out_path , 'my_new_video.mp4')
clip2 = ImageSequenceClip(new_paths , fps = 10)
clip2.write_videofile(out_video2)
```

```python
# Now we can also convert an image into an frame to make video more correct using ImageClip :

my_new_path = []
for path in list(new_paths):
    frame = ImageClip(path)
    print(frame.img)
#     print(dir(frame))
    my_new_path.append(frame.img)
    
# Now just give the path for the output video (the new one):
out_video3 = os.path.join(out_path , 'my_new_video2.mp4')
clip2 = ImageSequenceClip(my_new_path , fps = 10)
clip2.write_videofile(out_video3)
```
